### 3. Bliższy podgląd danych
Do opisanego później testu dwóch populacji postanowiłem użyć dwóch statystyk, **Happiness Score** oraz **Health (Life Expectancy)**, dlatego chciałbym się im przyjrzeć bliżej.

```python
regions = df_2016['Region'].unique()
def get_column_by_region(column_name):
    results = list()
    for region in regions:
        r = df_2016[df_2016['Region'] == region]
        x = sum(r[column_name]) / len(r)
        results.append(x)
            
    df_by_region = pd.DataFrame({'Region': regions, column_name: results})
    sorted_index = (df_by_region[column_name].sort_values(ascending=False)).index.values
    return df_by_region.reindex(sorted_index)
```

#### 3.1 Happiness Score wg regionów
Dane na temat radości z życia pogrupowałem wg regionów z którego dane państwo pochodzi. Jak widać na wykresie "najlepiej" żyje się w dobrze rozwiniętych państwach, natomiast najbiedniejsze regiony znacznie odstają od czołówki.

```python
df_happiness_by_region = get_column_by_region('Happiness Score')

plt.figure(figsize=(10, 8))
sns.barplot(x=df_happiness_by_region['Region'], y=df_happiness_by_region['Happiness Score'], 
            palette=sns.color_palette('RdBu_r', len(df_happiness_by_region)))
plt.xticks(rotation=90)
plt.show()
```

#### 3.2 Przewidywana długość życia wg regionów
Bardzo podobne wnioski można wyciągnąć z danych o przewidywanej długości życia. Wyraźnie też widać, że mimo tego że kolejność regionów jest bardzo podobna do tej z wykresu o radości z życia, to różnice między poszczególnymi lokalizacjami są dużo większe.

```python
df_health_by_region = get_column_by_region('Health (Life Expectancy)')

plt.figure(figsize=(10, 8))
sns.barplot(x=df_health_by_region['Region'], y=df_health_by_region['Health (Life Expectancy)'], 
            palette=sns.diverging_palette(145, 280, s=85, l=25, n=len(regions)))
plt.xticks(rotation=90)
plt.show()
```

```python
df_happiness_by_region_normalized = df_happiness_by_region
df_min = df_happiness_by_region_normalized['Happiness Score'].min()
df_max = df_happiness_by_region_normalized['Happiness Score'].max()
df_happiness_by_region_normalized['Happiness Score'] = (df_happiness_by_region_normalized['Happiness Score'] - df_min) / (df_max - df_min)

df_health_by_region_normalized = df_health_by_region
df_min = df_health_by_region_normalized['Health (Life Expectancy)'].min()
df_max = df_health_by_region_normalized['Health (Life Expectancy)'].max()
df_health_by_region_normalized['Health (Life Expectancy)'] = (df_health_by_region_normalized['Health (Life Expectancy)'] - df_min) / (df_max - df_min)
df_health_by_region_normalized = df_health_by_region_normalized.reindex(df_happiness_by_region_normalized.index.values)
```

#### 3.3 Porównanie radości z życia i przewidywanej długości życia
Do narysowania tego wykresu potrzebna była normalizacja obu statystyk, widać że obie wartości są skorelowane pomimo kilku skoków wykresu.

```python
plt.figure(figsize=(10,8))
plt.plot(df_happiness_by_region_normalized['Region'], df_happiness_by_region_normalized['Happiness Score'], '.-',
         color="#9b59b6", linewidth=2, markersize=12)
plt.plot(df_health_by_region_normalized['Region'], df_health_by_region_normalized['Health (Life Expectancy)'], '.-', 
         color="#2ecc71", linewidth=2, markersize=12)
plt.xticks(rotation=90)
plt.title('Happiness Score VS Life Expectancy')
plt.xlabel('Regions')
plt.ylabel('Normalized Values')
plt.legend()
plt.show()
```

### 5. Test sparowany
Do przeprowadzenia tego testu zostały użyte dwie statystyki - **Happiness Score** i **Health (Life Expectancy)**. Hipotezą zerową jest założenie, że róznica między obiema znormalizowanymi statystykami jest równa 0. Przperowadzono test dwustronny z wartością krytyczną $\alpha$ = 0.05 

```python
df_test = df_2016.loc[:, ['Country', 'Happiness Score', 'Health (Life Expectancy)']]
df_ha_max = df_test['Happiness Score'].max()
df_ha_min = df_test['Happiness Score'].min()

df_he_max = df_test['Health (Life Expectancy)'].max()
df_he_min = df_test['Health (Life Expectancy)'].min()

df_test['Happiness Score'] = (df_test['Happiness Score'] - df_ha_min) / (df_ha_max - df_ha_min)
df_test['Health (Life Expectancy)'] = (df_test['Health (Life Expectancy)'] - df_he_min) / (df_he_max - df_he_min)

df_test.describe()
```

#### 5.1 Wykres skrzypcowy różnicy
Z wykresu można odczytać, że wartości nie są zbyt oddalone od średniej, co udowadnia korelację tych statystyk

```python
df_test['Difference'] = df_test['Happiness Score'] - df_test['Health (Life Expectancy)']
sns.violinplot(y='Difference', data=df_test, inner="quartile")
plt.show()
```

```python
h_0 = 0
mean = df_test['Difference'].mean()
std = df_test['Difference'].std()
n = len(df_test)
T = (mean - h_0) / std * math.sqrt(n)
T
```

### 6. Podsumowanie testu
Wartość statystyki testowej wynosi **$T = -3.68$** <br>
Wartośc krytyczna wynosi $t_{kr}$ = $\theta^{-1}(1 - \frac{\alpha}{2}) = 1,96$
